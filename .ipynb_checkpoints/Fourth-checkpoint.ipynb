```python
def prepare_data_for_predition(df):
    df['Speed (m/ps)'] = df.apply(add_wind_speed, axis=1)
    df['Direction (degrees)'] = df.apply(wind_radians_degrees, axis=1)
    df['Direction (sector)'] = df.apply(direction_sector, axis=1)
    return df
```

```python
prepped_df = prepare_data_for_predition(pred_df)
prepped_df
```

```python
model_mapping = {
    'North': 0,
    'East': 1,
    'South': 2,
    'West': 3
}
```

```python
def make_prediction(row, models):
    model_mapping = {
        'North': 0,
        'East': 1,
        'South': 2,
        'West': 3
    }
    direction = row['Direction (sector)']
    model_index = model_mapping.get(direction)
    model = models[model_index]

    # Reshape the input data X
    X = row['Speed (m/ps)'].reshape(-1, 1)

    y_pred = model.predict(X)
    return y_pred
```

```python
prepped_df['pred'] = prepped_df.apply(make_prediction, axis=1, models=models)
prepped_df.head()
```
